```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ["sepal_len", "sepal_wid", "petal_len", "petal_wid", "class"]
iris_data = pd.read_csv(url, names=column_names)
```

```python
sns.pairplot(iris_data, hue="class")
plt.show()
```

```python
A = iris_data.drop("class", axis=1)
A
```

```python
B = iris_data["class"]
B
```

```python
A_train, A_test, B_train, B_test = train_test_split(A, B, test_size=0.3, random_state=50)
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(A_train, B_train)
```

```python
B_pred = knn.predict(A_test)
```

```python
print("Accuracy:", accuracy_score(B_test, B_pred))
```

```python
print(classification_report(B_test, B_pred))
```
